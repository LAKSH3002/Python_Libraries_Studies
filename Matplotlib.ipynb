```python
# Cheat sheet of Matplotlib

# 1) plotting
plt.plot() - Plots a line chart between two variables
plt.show() - displays the graph
plt.xlabel() - provides a label to the x-axis
plt.ylabel() - provides a label to the y-axis
plt.title() - providing a title to the graph
plt.grid() - adding grid lines to the graph
We can design all the above labels and grids as we want.

plt.subplot() - lets you create multiple graphs in the same plot, just position needs to be assigned.
plt.suptitle() - provides a title to the subplot.

# 2) Line chart elements
Markers: o (circle), *(star), .(point) ,x (X), X (filled X), +(Plus), P(plus filled), s(Square)
ms: marker size
mec = marker edge colour
mfc: marker face colour

# 3) Different graphs
plt.scatter(var1, var2, colors(optional)) - Scatter plot (best for continous variable)
plt.bar(var1, var2) - bar chart
plt.barh(var1, var2) - horizontal bar chart
plt.hist(histogram variable) - creates a histogram
plt.pie() - creates a pie chart, we can add labels, colours, explode (to make one edge stand out), startangle, legend, legend header.
```

```python
# MATPLOTLIB REVISION
'''
Low level graph plotting library in python
Mostly written in python, a few segments are written in C, JS for platform compatibility.
Platforms to practice python: Colab, Jupyter, anaconda, Spyder
'''
import matplotlib
print(matplotlib.__version__)

'''
Plotting
Markers
Line
Labels, Grids
Subplot,
scatterplot
Bars
Histograms
'''
```

```python
# Most of the utilities of matplotlib lies under pyplot (sub module) and are usually imported under the plt alias
import matplotlib.pyplot as plt
import numpy as np
x = np.array([0,6,10,15])
y = np.array([0,250,27,91])

# By default the plot functions created a line
# And there is a need to display x and y axis.
# marker is to emphasize on a specific point.
plt.plot(x, y)
plt.show()
```

```python
# Markers and its use in plotting the graph:
# o (circle), *(star), .(point) ,x (X), X (filled X), +(Plus), P(plus filled), s(Square)
# d (thin diamond), D (Diamond), v (Triangle down), ^ (Triangle up)
# 1 (Tri Down), 2 (Tri Up), 3 (Tri left), 4 (Tri right), | (V Line), _(Hline), p (pentagon), h & H (hexagon)

# color reference
# r, g, b, c, m - magenta, y, k - black, w - white.

# marker size: ms=20
# Marker color:
# Marker edge color >> mec = 'r'
# Marker Face color >> mfc = 'r'
print()
x2 = np.array([10,20,30,40,50,60])
y2 = np.array([3,8,1,12,18,26])
plt.plot(x2, y2, marker = 'o', ms=20, mec='r', mfc='y') # you can also use hexadecimal colour codes, hotpink
plt.show()
```

```python
# Lines in Matplotlib
# Format strings FMT
# '-' solid line, ':' dotted line, '--' dashed line, '-.' dashed/dotted line
# Use linestyle/ls to give style to a line, color using c or color, linewidth
x3 = np.array([10,20,30,40,50,60])
y3 = np.array([3,8,1,12,7,22])
y4 = np.array([2,9,7,11,12,21])
plt.plot(x3,y3,linestyle='dotted', linewidth=3, color= 'y')
plt.show()

# we can plot multiple lines in the same graph as well.
print()
plt.plot(y3)
plt.plot(y4)
plt.show()
```

```python
# Labelling the graph
# x-axis labelling, y-axis labelling and title labelling of the graph.
# you can style the labels using color, size and loc
x4 = np.array([10,20,30,40,50,60])
y4 = np.array([3,8,3,12,7,22])

plt.plot(x4, y4, 'o', ms=20, mec='y', mfc='y', color='r', linestyle='dashed') # markers and its attributes
plt.xlabel("Age of Cars", color='b') # x axis label with color
plt.ylabel("Average Speed", size=20) # y axis label with size
plt.title("CAR PLAY", loc='left') # Title along with location
plt.grid(color='green', linestyle="solid", linewidth='1', axis='y') # creating a grid of the graph
plt.show()
```

```python
# The subplot() function takes three arguments that describes the layout of the figure.
# The layout is organized in rows and columns, which are represented by the first and second argument.
# the third argument takes in order/sequence of the plots to be displayed.
# plot 1
x5 = np.array([0, 1, 2, 3, 4, 5])
y5 = np.array([10, 40, 50, 25, 45, 60])

# 1 row, 2 columns, 1st plot
plt.subplot(2, 2, 1)
plt.plot(x5, y5)
plt.title('marketing')

# plot 2
x6 = np.array([0, 1, 2, 3, 4, 5])
y6 = np.array([3, 8, 1, 10, 20, 23])

plt.subplot(2, 2, 2)
plt.plot(x6, y6)
plt.title("sales") # Title of plot number 2

# plot 3
x7 = np.array([46, 21, 42, 33, 34,])
y7 = np.array([50, 90, 70, 50, 50])

# 1 row, 2 columns, 2nd plot
plt.subplot(2, 2, 3)
plt.plot(x7, y7, color='red') # can give hexadecimal value and hotpink

# plot 4
x8 = np.array([5,7,8,7,2,17,2,9,4])
y8 = np.array([99,86,87,88,111,86,103,87,94])
# You can color each dot as well, by providing an array of colors
colors = np.array(["red", "green", "blue", "pink", "black", "yellow"])

plt.subplot(2, 2, 4)
plt.plot(x8, y8)

plt.suptitle("Fashion Designing")
plt.show()

# subplot(x,y,Position in the matrix)
# x - 1st dimension of the matrix
# y - 2nd dimension of the matrix
# plt.suptitle() for the subplot.
```

```python
# Scatter plot using scatter() function
# needs two arrays of the same length

# Scatterplots are generally used to find relationships between numerical values.
# We can provide color for each dot manually in case of small arrays, for large arrays we generally avoid
# plot 1
x9 = np.array([46, 21, 42, 33, 34, 15,32,41])
y9 = np.array([50, 90, 70, 50, 50, 100,47, 63])

# 1 row, 2 columns, 2nd plot
plt.subplot(1, 2, 1)
plt.scatter(x9, y9, color='red') # can give hexadecimal value and hotpink

# plot 2
x10 = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y10 = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
# You can color each dot as well, by providing an array of colors
colors = np.array(["red", "green", "blue", "pink", "black", "yellow"])

plt.subplot(1, 2, 2)
# we can use cmap function as well
# colormap is a list of colors.
# Different colormaps - accent, blues, brbg, BuGn
plt.scatter(x10, y10, cmap='viridis')
plt.colorbar()

plt.suptitle("Fashion Designing")
plt.show()

# We can also provide sizes and transparency for each dots in the scatterplot.
```

```python
# Bar chart
x_bar = np.array(["A","B","C","D","E", "F", "G", "H", "I"])
y_bar = np.array([3,7,5,12,22,13,11,4,3])

# bar() function takes arguments that describes the layout of the bars.
# we can provide colors, height, width too.
# default width and height of bars is 0.8
plt.bar(x_bar, y_bar, color='red', width=0.2)
plt.show()
plt.barh(x_bar, y_bar, color='red', height=0.2)
plt.show()

# For horizontal bars: plt.barh (x,y, color='xcode', height=0.1)
# color can be with hex codes and names
# Default height and width of the bars are 0.8
```

```python
# Histogram - A graph showing frequency distributions
#  A graphs showing the number of observations within each given interval.
histo = np.random.normal(170,10,150)
plt.hist(histo, color='red', width=1.2) # we will have to check for default value for width in histogram.
plt.show()
```

```python
# Pie charts
# labels, start angles, explode, shadow, colors, legend
p = np.array([35,25,15,10])
# labels
mylabels =  ["Apples", "Bananas", "Cherries", "Dates"]
# Default start angle is at the x-axis, but you can use startangle to give it a start
plt.pie(p, labels = mylabels, startangle = 90)
plt.show()

p2 = np.array([35,25,25,17])
mylabels2 = ["Car","Bike","Truck","Bus"]
# Explode: This will make one of the wedges stand out. We can also add shadow
myexplode = [0.2,0,0,0]

print()
plt.pie(p2, labels = mylabels2, explode = myexplode, shadow= True)
plt.show()

# Another thing that we can do is add colors as per our preferences by creating an array and passing that in pie() function.
# start angles
# As mentioned the default start angle is at the x-axis, but you can change the start angle by specifying a start angle parameter.
# The start angle parameter is defined with an angle in degrees, default angle is 0

# Explode
# It helps one of the parameter to stand out.
# The explode parameter, if specified, and not None, must be an array with one value for each wedge

# Shadow
# Add a shadow to the pie chart by setting the shadows parameter to True:
# You can also customize the colours of the chart by making an array and passing it in the function

# Legend
# To add a list of explanation for each wedge, use the legend() function:

# Legend Header
# To add a header to the legend, add the title parameter to the legend function
print()
z = np.array([35, 25, 28, 12])
my_labels = ["apples", "Fruits", "Veggies", "Olives"]
my_explode = [0, 0.2, 0, 0]
plt.pie(z, labels=my_labels, explode=my_explode, shadow=True)
plt.legend(title="Four Things:")
plt.show()
```
